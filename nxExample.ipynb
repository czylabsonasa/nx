## generate a small instance of the $G(n,m)$ random graph (Erdos-Renyi)

```python
import matplotlib.pyplot as plt
from networkx import nx

n = 100  # 10 nodes
m = 180  # 20 edges

G = nx.gnm_random_graph(n, m, seed=32) # reproducible

# some properties
# print("node degree clustering")
# for v in nx.nodes(G):
#     print('%s %d %f' % (v, nx.degree(G, v), nx.clustering(G, v)))
nx.draw(G,with_labels=True)
plt.show()
```

## read-write adjlist

```python
nx.write_adjlist(G,"g1_adj.txt")
H=nx.read_adjlist("g1_adj.txt")
#nx.draw(H)
nx.is_isomorphic(G,H)
```

## k-core

```python
kc=nx.core_number(G)
#print(kc)
s=[kc[i] for i in range(nx.number_of_nodes(G))]
ms=max(s)
slist=[ [] for i in range(ms+1) ]
#print(slist)
for i in range(nx.number_of_nodes(G)):
  slist[kc[i]].append(i)
#print(slist)  
slist=[v for v in slist if len(v)>0 ]
slist.reverse()
#print(slist)
nx.draw_spring(G, node_size=10)
```
