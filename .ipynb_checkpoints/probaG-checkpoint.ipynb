```python
import networkx as nx
import matplotlib.pyplot as plt
import random as rnd
```

```python
G=nx.barabasi_albert_graph(20,3)
#print(G.edges)
nx.draw(G)
```

```python
s=[[],[]]

for v in G.nodes:
  x=0
  if rnd.random()>0.5:
    x=1
  G.nodes[v]["shell"]=x
  s[x]+=[v]

#print(G.nodes.data())  
nx.draw_shell(G,nlist=s)
```
