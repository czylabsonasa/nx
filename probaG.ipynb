```python
import networkx as nx
import matplotlib.pyplot as plt
import random as rnd
```

```python
G=nx.barabasi_albert_graph(20,3)
print(G.number_of_nodes())
nx.draw(G)
help(nx.shortest_path_length)
```

```python
color=["red","blue"]
s=[[],[]]
col=[]
for v in G.nodes:
  x=rnd.choice([0,1])
  s[x]+=[v]
  G.nodes[v]["color"]=color[x]
#print(G.nodes.data())  
  
nx.draw_shell(G,nlist=s,node_color=[G.nodes[v]["color"] for v in G.nodes ])
```
